```python
#Finding the percentage of missing values
for column in data.columns:
    percentage = data[column].isnull().mean()
    print(f'{column}: {round(percentage*100, 2)}%')
```

```python
#Handling null values. Here we are removing the observations where the director column is empty
no_director = data[data['Director'].isnull()].index
no_director
data.drop(no_director, axis=0,inplace=True)
data
#After dropping the null values, we get 5401 rows and 11 columns
```

```python
#Replacing the null value of rating by average value
mode = ''.join(data['Rating'].mode())
data['Rating'].fillna(mode, inplace=True)
data
```

```python
#Checking for other missing values
data.isnull().sum()
```

```python
#Dropping the cast observations where values are null
no_cast = data[data['Cast'].isnull()].index
data.drop(no_cast, axis=0,inplace=True)
data
```

```python
#Filling the missing country details with United States
data['Country'].fillna('United States', inplace=True)
data['Country'].isnull().any()
```

```python
#Now we got the clean dataset.
data.isnull().sum()
```

```python
#checking for any duplicate values
data[data.duplicated()]
```

```python
#Removing the duplicated values permanently
data.drop_duplicates(inplace=True)
#After removing the duplicate values, the dataset will have 7787 rows and 11 columns
```

```python
#Once again confirming for duplicate values
data[data.duplicated()]
#Now there are no duplicate values in the dataset
```

```python
#Encoding categorical values. here we are encoding the Category column which has 2 values, TV shows and Movie.
from sklearn.preprocessing import LabelEncoder
```

```python
#We are transforming the Category column
l3=LabelEncoder()
label=l3.fit_transform(data['Category'])
```

```python
#To see the classes of category column, which contain 2 values Movie and TV Show
l3.classes_
```

```python
#we are dropping the existing category column inorder to append the new transformed column
new=data.drop("Category",axis='columns')
```

```python
#The new dataset which does not contain the category column
new
```

```python
#We are appending the new column Category with label encoded values
new["Category"]=label
```
